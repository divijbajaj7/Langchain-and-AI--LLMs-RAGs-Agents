# Pattern 1: LLM Responder

## What is it?
The simplest pattern - direct interaction with an LLM. The LLM receives a prompt and generates a response.

## Component Focus: **LLM**
This pattern highlights the core LLM capability - understanding context and generating relevant responses.

## Use Cases
- Chatbots
- Content generation
- Question answering
- Translation

````python
# Pattern 1: Simple LLM Responder

class LLMResponder:
    """A simple LLM-based responder that processes user queries."""
    
    def __init__(self, system_prompt: str = "You are a helpful assistant."):
        self.system_prompt = system_prompt
        self.client = OpenAI()
    
    def respond(self, user_message: str, model: str = "gpt-4o-mini") -> str:
        """Generate a response to the user's message."""
        response = self.client.chat.completions.create(
            model=model,
            messages=[
                {"role": "system", "content": self.system_prompt},
                {"role": "user", "content": user_message}
            ],
            temperature=0.7
        )
        return response.choices[0].message.content

# Create different responders with different personas
general_assistant = LLMResponder("You are a helpful general assistant.")
code_expert = LLMResponder("You are an expert Python programmer. Provide concise code solutions.")
creative_writer = LLMResponder("You are a creative writer. Be imaginative and engaging.")
````

# Pattern 2: Agentic Router

## What is it?
The LLM acts as a decision-maker, routing requests to different handlers based on the input. This introduces **Planning** - the agent decides what action to take.

## Component Focus: **Planning**
The LLM analyzes the input and makes a decision about how to handle it.

## Use Cases
- Customer support triage
- Intent classification
- Multi-domain assistants
- Workflow automation

````python
# Pattern 2: Agentic Router using Structured Outputs

# Define the routing decision schema
class RoutingDecision(BaseModel):
    """The router's decision on how to handle a request."""
    category: Literal["technical", "billing", "general", "complaint"] = Field(
        description="The category of the user's request"
    )
    confidence: float = Field(
        description="Confidence score between 0 and 1",
        ge=0, le=1
    )
    reasoning: str = Field(
        description="Brief explanation of why this category was chosen"
    )

class AgenticRouter:
    """Routes user requests to appropriate handlers based on content analysis."""
    
    def __init__(self):
        self.client = OpenAI()
        # Define specialized handlers for each category
        self.handlers = {
            "technical": LLMResponder("You are a technical support specialist. Help with technical issues."),
            "billing": LLMResponder("You are a billing specialist. Help with payment and subscription queries."),
            "general": LLMResponder("You are a general support agent. Provide helpful information."),
            "complaint": LLMResponder("You are a customer success manager. Be empathetic and solution-focused.")
        }
    
    def route(self, user_message: str) -> RoutingDecision:
        """Analyze the message and decide which handler to use."""
        response = self.client.beta.chat.completions.parse(
            model="gpt-4o-mini",
            messages=[
                {
                    "role": "system",
                    "content": """You are a routing agent. Analyze the user's message and categorize it:
                    - technical: Software bugs, how-to questions, technical issues
                    - billing: Payment issues, subscription, pricing questions
                    - general: General inquiries, information requests
                    - complaint: Frustration, dissatisfaction, complaints
                    
                    It can have multiple catrgories as well. If you think user has issues with 2 or more problems"""
                },
                {"role": "user", "content": user_message}
            ],
            response_format=RoutingDecision
        )
        return response.choices[0].message.parsed
    
    def handle(self, user_message: str) -> dict:
        """Route the message and generate a response."""
        # Step 1: Plan - Decide where to route
        decision = self.route(user_message)
        
        # Step 2: Execute - Use the appropriate handler
        handler = self.handlers[decision.category]
        response = handler.respond(user_message)
        
        return {
            "routing": {
                "category": decision.category,
                "confidence": decision.confidence,
                "reasoning": decision.reasoning
            },
            "response": response
        }
````

````python
# Test the Agentic Router
router = AgenticRouter()

test_messages = [
    "My app keeps crashing when I try to upload files",
    "I was charged twice for my subscription this month!",
    "What are your business hours?",
    "This is ridiculous! I've been waiting for 3 days for a response!",
    "My app is crashing and I'm facing issues with invoicing as well"
]

for message in test_messages:
    print("=" * 60)
    print(f"USER: {message}")
    print("-" * 60)
    
    result = router.handle(message)
    
    print(f"ROUTED TO: {result['routing']['category']} (confidence: {result['routing']['confidence']:.2f})")
    print(f"REASONING: {result['routing']['reasoning']}")
    print(f"\nRESPONSE: {result['response'][:200]}...")
    print()
````

### Key Takeaways - Pattern 2
- **Planning** is introduced - the agent decides what to do
- Structured outputs ensure reliable routing decisions
- Different handlers can have specialized knowledge
- The router is the "orchestrator" that coordinates the flow

# Pattern 3: Tool-Using Agent

## What is it?
An agent that can use external **Tools** to accomplish tasks. The LLM decides when to use tools and interprets their results.

## Component Focus: **Tools**
Tools extend the agent's capabilities beyond text generation - calculators, APIs, databases, etc.

## Use Cases
- Data retrieval and analysis
- Web search assistants
- Code execution
- API integrations

````python
# Pattern 3: Tool-Using Agent

import math
from datetime import datetime
import requests

# Define our tools (functions the agent can call)
def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        # Safe evaluation of mathematical expressions
        allowed_names = {"sin": math.sin, "cos": math.cos, "tan": math.tan,
                        "sqrt": math.sqrt, "pi": math.pi, "e": math.e,
                        "log": math.log, "pow": pow, "abs": abs}
        result = eval(expression, {"__builtins__": {}}, allowed_names)
        return f"Result: {result}"
    except Exception as e:
        return f"Error: {str(e)}"

def get_current_time() -> str:
    """Get the current date and time."""
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

def google_search( query=""):
    """Simulate a Google search (placeholder)."""
    # To install: pip install tavily-python
    from tavily import TavilyClient
    client = TavilyClient("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")
    response = client.search(
        query="",
        search_depth="advanced"
    )
    print(response)


def get_weather(city: str) -> str:
    """Get weather information for a city (simulated)."""
    # Simulated weather data (in production, you'd call a real API)
    weather_data = {
        "new york": {"temp": 72, "condition": "Sunny", "humidity": 45},
        "london": {"temp": 59, "condition": "Cloudy", "humidity": 78},
        "tokyo": {"temp": 68, "condition": "Partly Cloudy", "humidity": 60},
        "paris": {"temp": 64, "condition": "Rainy", "humidity": 85},
    }
    city_lower = city.lower()
    if city_lower in weather_data:
        w = weather_data[city_lower]
        return f"Weather in {city}: {w['temp']}°F, {w['condition']}, Humidity: {w['humidity']}%"
    return f"Weather data not available for {city}"

def search_wikipedia(query: str) -> str:
    """Search Wikipedia for information (simplified)."""
    try:
        url = f"https://en.wikipedia.org/api/rest_v1/page/summary/{query.replace(' ', '_')}"
        response = requests.get(url, timeout=5)
        if response.status_code == 200:
            data = response.json()
            return data.get("extract", "No information found.")[:500]
        return "Could not find information on Wikipedia."
    except Exception as e:
        return f"Search error: {str(e)}"

# Tool registry with descriptions for the LLM
TOOLS = [
    {
        "type": "function",
        "function": {
            "name": "calculator",
            "description": "Evaluate mathematical expressions. Supports basic arithmetic, trigonometry (sin, cos, tan), sqrt, log, pi, e.",
            "parameters": {
                "type": "object",
                "properties": {
                    "expression": {
                        "type": "string",
                        "description": "The mathematical expression to evaluate, e.g., '2 + 2' or 'sqrt(16)'"
                    }
                },
                "required": ["expression"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "get_current_time",
            "description": "Get the current date and time.",
            "parameters": {
                "type": "object",
                "properties": {},
                "required": []
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "get_weather",
            "description": "Get current weather information for a city.",
            "parameters": {
                "type": "object",
                "properties": {
                    "city": {
                        "type": "string",
                        "description": "The name of the city"
                    }
                },
                "required": ["city"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "search_wikipedia",
            "description": "Search Wikipedia for information about a topic.",
            "parameters": {
                "type": "object",
                "properties": {
                    "query": {
                        "type": "string",
                        "description": "The topic to search for"
                    }
                },
                "required": ["query"]
            }
        }
    }
]

# Function dispatcher
TOOL_FUNCTIONS = {
    "calculator": calculator,
    "get_current_time": get_current_time,
    "get_weather": get_weather,
    "search_wikipedia": search_wikipedia
}
````

````python
# The Tool-Using Agent

class ToolAgent:
    """An agent that can use tools to accomplish tasks."""
    
    def __init__(self):
        self.client = OpenAI()
        self.tools = TOOLS
        self.tool_functions = TOOL_FUNCTIONS
    
    def run(self, user_message: str, verbose: bool = True) -> str:
        """Process a user message, using tools as needed."""
        messages = [
            {
                "role": "system",
                "content": """You are a helpful assistant with access to tools.
                Use tools when needed to provide accurate information.
                Always explain what you're doing and interpret tool results for the user."""
            },
            {"role": "user", "content": user_message}
        ]
        
        while True:
            # Get LLM response
            response = self.client.chat.completions.create(
                model="gpt-4o-mini",
                messages=messages,
                tools=self.tools,
                tool_choice="auto"
            )
            
            assistant_message = response.choices[0].message
            
            # Check if we need to call tools
            if assistant_message.tool_calls:
                messages.append(assistant_message)
                
                # Execute each tool call
                for tool_call in assistant_message.tool_calls:
                    function_name = tool_call.function.name
                    function_args = json.loads(tool_call.function.arguments)
                    
                    if verbose:
                        print(f"🔧 TOOL CALL: {function_name}({function_args})")
                    
                    # Execute the tool
                    if function_name in self.tool_functions:
                        result = self.tool_functions[function_name](**function_args)
                    else:
                        result = f"Unknown tool: {function_name}"
                    
                    if verbose:
                        print(f"📋 RESULT: {result}\n")
                    
                    # Add tool result to messages
                    messages.append({
                        "role": "tool",
                        "tool_call_id": tool_call.id,
                        "content": result
                    })
            else:
                # No more tool calls, return final response
                return assistant_message.content
````

# Pattern 4: Multi-Agent System

## What is it?
Multiple specialized agents working together, each with specific expertise. An orchestrator coordinates their work.

## Component Focus: **Planning + Coordination**
Complex planning across multiple agents, each contributing their specialty.

## Use Cases
- Research teams (researcher, writer, editor)
- Software development (architect, developer, tester)
- Content pipelines (ideation, creation, review)

````python
# Pattern 4: Multi-Agent System

class SpecializedAgent:
    """A specialized agent with a specific role and expertise."""
    
    def __init__(self, name: str, role: str, system_prompt: str):
        self.name = name
        self.role = role
        self.system_prompt = system_prompt
        self.client = OpenAI()
    
    def process(self, task: str, context: str = "") -> str:
        """Process a task with optional context from other agents."""
        messages = [
            {"role": "system", "content": self.system_prompt},
            {"role": "user", "content": f"Context from team:\n{context}\n\nYour task:\n{task}" if context else task}
        ]
        
        response = self.client.chat.completions.create(
            model="gpt-4o-mini",
            messages=messages,
            temperature=0.7
        )
        return response.choices[0].message.content


class ContentCreationTeam:
    """A multi-agent team for content creation."""
    
    def __init__(self):
        # Create specialized agents
        self.researcher = SpecializedAgent(
            name="Researcher",
            role="Research Specialist",
            system_prompt="""You are a research specialist. Your job is to:
            - Identify key topics and angles for a given subject
            - Provide factual information and data points
            - Suggest relevant examples and case studies
            Be thorough but concise. Output structured research notes."""
        )
        
        self.writer = SpecializedAgent(
            name="Writer",
            role="Content Writer",
            system_prompt="""You are a professional content writer. Your job is to:
            - Transform research into engaging content
            - Write in a clear, compelling style
            - Structure content logically with good flow
            Use the research provided to create polished content."""
        )
        
        self.editor = SpecializedAgent(
            name="Editor",
            role="Editor & Quality Assurer",
            system_prompt="""You are a senior editor. Your job is to:
            - Review and improve content quality
            - Check for clarity, coherence, and engagement
            - Suggest improvements and polish the final output
            - Provide the final, publication-ready version
            Be constructive and output the improved content."""
        )
        
        self.client = OpenAI()
    
    def create_content(self, topic: str, content_type: str = "blog post", verbose: bool = True) -> dict:
        """Orchestrate the team to create content."""
        results = {}
        
        # Step 1: Research
        if verbose:
            print("📚 RESEARCHER is gathering information...")
        research = self.researcher.process(
            f"Research the following topic for a {content_type}: {topic}\n"
            "Provide key facts, interesting angles, and useful examples."
        )
        results["research"] = research
        if verbose:
            print(f"Research complete!\n")
        
        # Step 2: Writing
        if verbose:
            print("✍️ WRITER is creating content...")
        draft = self.writer.process(
            f"Write a {content_type} about: {topic}",
            context=f"Research notes:\n{research}"
        )
        results["draft"] = draft
        if verbose:
            print(f"Draft complete!\n")
        
        # Step 3: Editing
        if verbose:
            print("📝 EDITOR is polishing the content...")
        final = self.editor.process(
            "Review and improve this content. Output the final polished version.",
            context=f"Draft to edit:\n{draft}"
        )
        results["final"] = final
        if verbose:
            print(f"Editing complete!\n")
        
        return results
````

````python
# Test the Multi-Agent Content Team
team = ContentCreationTeam()

results = team.create_content(
    topic="The benefits of learning to code in 2026",
    content_type="short blog post"
)

print("=" * 60)
print("FINAL CONTENT")
print("=" * 60)
print(results["final"])
````

# Pattern 5: Autonomous Agent with Memory

## What is it?
A self-directing agent that can remember past interactions, plan multi-step tasks, and work autonomously toward a goal.

## Component Focus: **Memory + Planning + Tools**
This pattern combines all components:
- **Memory**: Remembers conversation history and past actions
- **Planning**: Breaks down goals into steps
- **Tools**: Executes actions in the world

## Use Cases
- Personal assistants
- Research agents
- Task automation

````python
# Pattern 5: Autonomous Agent with Memory (ReAct Pattern)

class ConversationMemory:
    """Simple conversation memory that stores chat history."""
    
    def __init__(self, max_messages: int = 20):
        self.messages: List[dict] = []
        self.max_messages = max_messages
        self.summary: str = ""
    
    def add(self, role: str, content: str):
        """Add a message to memory."""
        self.messages.append({"role": role, "content": content})
        # Keep only recent messages
        if len(self.messages) > self.max_messages:
            self._summarize_old_messages()
    
    def _summarize_old_messages(self):
        """Summarize older messages to save context."""
        old_messages = self.messages[:10]
        self.messages = self.messages[10:]
        # In production, you'd use an LLM to summarize
        self.summary += f"\n[Previous conversation summary: {len(old_messages)} messages discussed]"
    
    def get_context(self) -> List[dict]:
        """Get the conversation context for the LLM."""
        context = []
        if self.summary:
            context.append({"role": "system", "content": f"Previous context: {self.summary}"})
        context.extend(self.messages)
        return context
    
    def clear(self):
        """Clear the memory."""
        self.messages = []
        self.summary = ""


class TaskMemory:
    """Memory for tracking tasks and their status."""
    
    def __init__(self):
        self.tasks: List[dict] = []
        self.completed: List[dict] = []
    
    def add_task(self, task: str, priority: int = 1):
        """Add a task to memory."""
        self.tasks.append({"task": task, "priority": priority, "status": "pending"})
    
    def complete_task(self, task: str, result: str):
        """Mark a task as complete."""
        for t in self.tasks:
            if t["task"] == task:
                t["status"] = "completed"
                t["result"] = result
                self.completed.append(t)
                self.tasks.remove(t)
                break
    
    def get_pending(self) -> List[dict]:
        """Get pending tasks."""
        return [t for t in self.tasks if t["status"] == "pending"]
    
    def get_summary(self) -> str:
        """Get a summary of task status."""
        pending = len(self.get_pending())
        completed = len(self.completed)
        return f"Tasks: {completed} completed, {pending} pending"
````

````python
# Autonomous Agent with ReAct Pattern

class AutonomousAgent:
    """An autonomous agent that can plan, act, and remember."""
    
    def __init__(self):
        self.client = OpenAI()
        self.conversation_memory = ConversationMemory()
        self.task_memory = TaskMemory()
        self.tools = TOOLS
        self.tool_functions = TOOL_FUNCTIONS
        self.max_iterations = 5
    
    def think(self, goal: str) -> dict:
        """Plan the steps needed to achieve a goal."""
        response = self.client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {
                    "role": "system",
                    "content": """You are a planning agent. Given a goal, break it down into concrete steps.
                    Output a JSON object with:
                    - "analysis": Brief analysis of what's needed
                    - "steps": Array of step strings
                    - "tools_needed": Array of tool names that might help"""
                },
                {"role": "user", "content": f"Plan how to achieve this goal: {goal}"}
            ],
            response_format={"type": "json_object"}
        )
        return json.loads(response.choices[0].message.content)
    
    def act(self, action: str) -> str:
        """Execute an action, potentially using tools."""
        messages = [
            {
                "role": "system",
                "content": "You are an execution agent. Complete the given task using available tools if needed."
            },
            {"role": "user", "content": action}
        ]
        
        response = self.client.chat.completions.create(
            model="gpt-4o-mini",
            messages=messages,
            tools=self.tools,
            tool_choice="auto"
        )
        
        assistant_message = response.choices[0].message
        
        if assistant_message.tool_calls:
            # Execute tools
            results = []
            for tool_call in assistant_message.tool_calls:
                func_name = tool_call.function.name
                func_args = json.loads(tool_call.function.arguments)
                if func_name in self.tool_functions:
                    result = self.tool_functions[func_name](**func_args)
                    results.append(f"{func_name}: {result}")
            return "\n".join(results)
        
        return assistant_message.content
    
    def reflect(self, goal: str, actions_taken: List[str], results: List[str]) -> str:
        """Reflect on progress and decide next steps."""
        response = self.client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {
                    "role": "system",
                    "content": """You are a reflection agent. Analyze progress toward a goal and provide insights.
                    Determine if the goal is achieved or what else needs to be done."""
                },
                {
                    "role": "user",
                    "content": f"""Goal: {goal}
                    
Actions taken:
{chr(10).join(f'- {a}' for a in actions_taken)}

Results:
{chr(10).join(f'- {r}' for r in results)}

Reflect on the progress. Is the goal achieved? What insights can you provide?"""
                }
            ]
        )
        return response.choices[0].message.content
    
    def run(self, goal: str, verbose: bool = True) -> dict:
        """Run the autonomous agent toward a goal."""
        if verbose:
            print(f"🎯 GOAL: {goal}\n")
        
        # Phase 1: Think/Plan
        if verbose:
            print("🧠 THINKING...")
        plan = self.think(goal)
        if verbose:
            print(f"Analysis: {plan.get('analysis', 'N/A')}")
            print(f"Steps: {plan.get('steps', [])}")
            print(f"Tools needed: {plan.get('tools_needed', [])}\n")
        
        # Add tasks to memory
        for step in plan.get("steps", []):
            self.task_memory.add_task(step)
        
        # Phase 2: Act
        actions_taken = []
        results = []
        
        for i, step in enumerate(plan.get("steps", []), 1):
            if i > self.max_iterations:
                break
                
            if verbose:
                print(f"⚡ ACTION {i}: {step}")
            
            result = self.act(step)
            actions_taken.append(step)
            results.append(result)
            
            if verbose:
                print(f"   Result: {result[:200]}...\n" if len(result) > 200 else f"   Result: {result}\n")
            
            # Update task memory
            self.task_memory.complete_task(step, result)
            
            # Store in conversation memory
            self.conversation_memory.add("assistant", f"Action: {step}\nResult: {result}")
        
        # Phase 3: Reflect
        if verbose:
            print("🔍 REFLECTING...")
        reflection = self.reflect(goal, actions_taken, results)
        if verbose:
            print(f"{reflection}\n")
        
        return {
            "goal": goal,
            "plan": plan,
            "actions": actions_taken,
            "results": results,
            "reflection": reflection,
            "task_summary": self.task_memory.get_summary()
        }
````

````python
# Test the Autonomous Agent
autonomous_agent = AutonomousAgent()

# Give it a complex goal
result = autonomous_agent.run(
    "I'm planning a trip to Paris. Help me gather information about the Eiffel Tower, "
    "check what time it is there (Paris is UTC+1), and calculate how many hours "
    "until 6 PM Paris time if it's currently morning."
)

print("=" * 60)
print(f"TASK SUMMARY: {result['task_summary']}")
````

````python
# Demonstrate Memory Persistence
print("\n" + "=" * 60)
print("MEMORY DEMONSTRATION")
print("=" * 60)

# The agent remembers previous interactions
print("\nConversation Memory:")
for msg in autonomous_agent.conversation_memory.get_context()[-3:]:
    print(f"  [{msg['role']}]: {msg['content'][:100]}...")

print("\nTask Memory:")
print(f"  Completed tasks: {len(autonomous_agent.task_memory.completed)}")
for task in autonomous_agent.task_memory.completed:
    print(f"    - {task['task']}")
````

### Key Takeaways - Pattern 5
- **Memory** allows agents to maintain context and learn from past interactions
- **Planning** (Think) breaks complex goals into manageable steps
- **Action** (Act) executes the plan using tools
- **Reflection** evaluates progress and provides insights
- The ReAct (Reasoning + Acting) loop enables autonomous behavior

# Bonus: Building with LangChain

LangChain provides higher-level abstractions for building agents. Here's how the same concepts map to LangChain components.

````python
# LangChain Implementation Example

from langchain_openai import ChatOpenAI
from langchain.tools import tool
from langchain.agents import AgentExecutor, create_react_agent
#from langchain.agents import AgentExecutor, create_react_agent
#from langchain.agents import tool, AgentExecutor, create_react_agent
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

# Initialize the LLM
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# Define tools using the @tool decorator
@tool
def calculate(expression: str) -> str:
    """Evaluate a mathematical expression. Example: calculate('2 + 2')"""
    try:
        result = eval(expression, {"__builtins__": {}}, 
                     {"sin": math.sin, "cos": math.cos, "sqrt": math.sqrt, "pi": math.pi})
        return str(result)
    except Exception as e:
        return f"Error: {e}"

@tool
def get_time() -> str:
    """Get the current date and time."""
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

@tool
def search_wiki(topic: str) -> str:
    """Search Wikipedia for information about a topic."""
    try:
        url = f"https://en.wikipedia.org/api/rest_v1/page/summary/{topic.replace(' ', '_')}"
        response = requests.get(url, timeout=5)
        if response.status_code == 200:
            return response.json().get("extract", "No info found")[:300]
        return "Topic not found"
    except:
        return "Search failed"

tools = [calculate, get_time, search_wiki]

print("LangChain tools defined!")
print(f"Available tools: {[t.name for t in tools]}")
````

````python
# Create a LangChain Agent with Memory

from langchain.agents import create_tool_calling_agent

# Create prompt with memory placeholder
prompt = ChatPromptTemplate.from_messages([
    ("system", """You are a helpful assistant with access to tools.
    Use tools when needed. Be concise in your responses."""),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

# Create memory
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)

# Create agent
agent = create_tool_calling_agent(llm, tools, prompt)

# Create executor with memory
agent_executor = AgentExecutor(
    agent=agent, 
    tools=tools, 
    memory=memory,
    verbose=True,
    handle_parsing_errors=True
)

print("LangChain agent created with memory!")
````

# Summary: AI Agent Components

| Component | Description | Patterns Used |
|-----------|-------------|---------------|
| **LLM** | The core reasoning engine | All patterns |
| **Planning** | Breaking down tasks, deciding actions | Router, Multi-Agent, Autonomous |
| **Tools** | External capabilities (APIs, functions) | Tool-Using, Autonomous |
| **Memory** | Conversation history, task tracking | Autonomous |

## Pattern Progression

1. **LLM Responder**: LLM only
2. **Agentic Router**: LLM + Planning (decision making)
3. **Tool-Using Agent**: LLM + Tools
4. **Multi-Agent**: LLM + Planning + Coordination
5. **Autonomous Agent**: LLM + Planning + Tools + Memory
